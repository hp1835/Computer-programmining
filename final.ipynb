```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('CO2Emissions_filtered.csv')
df.head()
```

```python
df['Country Code'] = df['Country Code'].str.lower()
df.head()
```

```python
co2_data = df.set_index('Country Code').T.to_dict('list')
print(co2_data)
```

```python
countries = ['dnk', 'fin', 'isl', 'nor', 'swe']
time_range = list(range(1960, 2015))
```

```python
nordic_df = df[df['Country Code'].isin(countries)]
nordic_emissions = nordic_df.set_index('Country Code').reindex(countries)
print("Columns in nordic_emissions DataFrame:", nordic_emissions.columns)
print(nordic_emissions.head())
```

```python
# Convert time_range years to strings
string_time_range = [str(year) for year in time_range]
print("String Time Range:", string_time_range)
```

```python

nordic_emissions = nordic_emissions.loc[:, string_time_range]
```

```python
print(nordic_emissions)
```

```python
import matplotlib.pyplot as plt
```

```python
# Convert years to strings
string_time_range = [str(year) for year in time_range]


plt.figure(figsize=(12, 6))

# Loop over countries 
for country in nordic_emissions.index:
    
    plt.plot(string_time_range, nordic_emissions.loc[country].values, label=country.upper())         # Plot emissions for each country


plt.title('CO2 Emissions (1960-2014) for Nordic Countries')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (tons)')
plt.xticks(rotation=45)
plt.legend(title='Country')

plt.grid()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

def smooth_a(a, window_size=11):
    """ Simple Moving Average with custom windowing """
    N = len(a)
    r = []
    
    for i in range(N):
        start = max(0, i - window_size // 2)
        end = min(N, i + window_size // 2 + 1)
        
        # Calculate the average of the window
        window_elements = a[start:end]
        r.append(sum(window_elements) / len(window_elements))
    
    return r

def smooth_b(a, span=11):
    """ Custom Exponential Moving Average """
    N = len(a)
    r = []
    
    # Initialize first element of EMA
    ema = a[0]  # Starting point for EMA
    
    for i in range(N):
        if i == 0:
            r.append(ema)
            continue
        
        alpha = 2 / (span + 1)  # Calculate smoothing factor
        ema = (a[i] * alpha) + (ema * (1 - alpha))  # Update EMA
        r.append(ema)
    
    return r
```

```python

colors = {
    'dnk': 'blue',
    'fin': 'green',
    'isl': 'orange',
    'nor': 'red',
    'swe': 'purple'
}
```

```python

# Plotting
plt.figure(figsize=(12, 6))

# Loop over each country in the DataFrame index
for country in nordic_emissions.index:
    emissions = nordic_emissions.loc[country].values  
    years = string_time_range  

    # Plot original data
    plt.plot(years, emissions, label=f"{country.upper()} - Original", color=colors[country], linestyle=':', alpha=0.7)

    # Apply smoothing 
    smoothed_a = smooth_a(emissions, window_size=11)
    smoothed_b = smooth_b(emissions, span=11)

    plt.plot(years, smoothed_a, label=f"{country.upper()} - Smooth A", color=colors[country])
    plt.plot(years, smoothed_b, label=f"{country.upper()} - Smooth B", color=colors[country], linestyle='--')

plt.title('CO2 Emissions (1960-2014) for Nordic Countries with Smoothing')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (tons)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend(title='Country and Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.tight_layout()
plt.show()
```
